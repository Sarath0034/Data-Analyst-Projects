```python
from IPython.display import clear_output
import random
board = [' '] * 10
computer = 'X'
human = 'O'


def display_board(board):
    print(f'{board[1]} | {board[2]} | {board[3]}')
    print(f'{board[4]} | {board[5]} | {board[6]}')
    print(f'{board[7]} | {board[8]} | {board[9]}')
    print('-' * 10)


def check_win():
    if board[1] == board[2] == board[3] and board[1] != ' ':
        return True
    elif board[4] == board[5] == board[6] and board[4] != ' ':
        return True
    elif board[7] == board[8] == board[9] and board[7] != ' ':
        return True
    elif board[1] == board[4] == board[7] and board[1] != ' ':
        return True
    elif board[2] == board[5] == board[8] and board[2] != ' ':
        return True
    elif board[3] == board[6] == board[9] and board[3] != ' ':
        return True
    elif board[1] == board[5] == board[9] and board[1] != ' ':
        return True
    elif board[7] == board[5] == board[3] and board[7] != ' ':
        return True
    else:
        return False


def check_draw():
    return board.count(' ') < 2  
def is_available(pos):
    return board[pos] == ' '


def insert(letter, pos):
    if is_available(pos):
        board[pos] = letter
        display_board(board)

        if check_win():
            print(f"{'Computer' if letter == 'X' else 'Human'} Wins!")
            exit()

        if check_draw():
            print("Draw")
            exit()
    else:
        print("Position occupied! Try again.")
        if letter == 'O':
            human_move()  
        else:
            computer_move()  


def human_move():
    pos = int(input("Enter the position to insert: "))
    while not (1 <= pos <= 9) or not is_available(pos):
        pos = int(input("Invalid or occupied! Enter a different position: "))
    insert(human, pos)
def computer_move():
    pos = random.choice([i for i in range(1, 10) if is_available(i)])
    insert(computer, pos)



while True:
    display_board(board)
    human_move()
    if check_win() or check_draw():
        break
    computer_move()
    if check_win() or check_draw():
        break
```
